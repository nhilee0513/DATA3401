```python
x=data5
y1=[]
b0=3
b1=2
for i in range(0,15):
    e=norm.rvs(0,10)
    y1.append(b0+x[i]*b1+e)


dat1={'x':x,
      'y':y1}
df1 = pd.DataFrame(dat1)
df1
```

```python
plt.scatter(x,y1)
```

```python
y2=[]

for i in range(0,15):
    e=norm.rvs(0,20)
    y2.append(b0+x[i]*b1+e)

dat2={'x':x,
      'y':y2}
df2= pd.DataFrame(dat2)
df2
```

```python
plt.scatter(x,y2)
```
