```python
# the usual starting block: 
import numpy as np 
%matplotlib inline   
import matplotlib.pyplot as plt
plt.style.use('fivethirtyeight')
import math
from scipy import stats
```

### matplotlib.pyplot.scatter() for plotting points

```python
y = stats.norm.rvs(size = 5)
x = np.arange(0,5)
plt.scatter(x, y)
```

```python
xvec = np.arange(-3,3, .1) #plot the normal distribution
plt.scatter(xvec, stats.norm.pdf(xvec),  c = 'blue')
```

### matplotlib.pyplot.plot() for plotting lines

```python
xvec = np.arange(-4,4, .01)
plt.plot(xvec, stats.norm.pdf(xvec), linewidth = 1, c = 'orangered')
```

### adding curves to plot

```python
xvec = np.arange(-7,9, .01)
plt.plot(xvec, stats.norm.pdf(xvec), linewidth = 1, c = 'orangered')
plt.plot(xvec, stats.norm.pdf(xvec, 1,2), linewidth = 1, c = 'crimson') # standard normal, have more variance, more scale
plt.plot(np.array([0,0]), np.array([0, stats.norm.pdf(0)]), c='orangered', linewidth = .5) # vertical line 
```
